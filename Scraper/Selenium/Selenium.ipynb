```python
#scroll down 2 times
#increase the range to sroll more
n_scrolls = 2
for j in range(0, n_scrolls):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(5)
```

```python
anchors = driver.find_elements_by_tag_name('a')
anchors = [a.get_attribute('href') for a in anchors]
#narrow down all links to image links only
anchors = [a for a in anchors if str(a).startswith("https://www.instagram.com/p/")]

print('Found ' + str(len(anchors)) + ' links to images')
anchors[:5]
```

```python
images = []

#follow each image link and extract only image at index=1
for a in anchors:
    driver.get(a)
    time.sleep(5)
    img = driver.find_elements_by_tag_name('img')
    img = [i.get_attribute('src') for i in img]
    images.append(img[1])
    
images[:5] 
```

```python
WebDriverWait(driver,10).until(
EC.text_to_be_present_in_element(
    (By.CLASS_NAME,"infinite-scroller__show-more-button"), #Element Filtration
    "See more jobs" #Expected text
)
)
```

```python
while 1:
    #time.sleep(5)
    show_more_Btn = driver.find_element_by_class_name("infinite-scroller__show-more-button")
    if show_more_Btn.text:
        driver.implicitly_wait(20)
        if show_more_Btn.is_enabled():
            driver.implicitly_wait(10)
            show_more_Btn.click()
        else:
            driver.implicitly_wait(20)
            print("break")
            break
    else:
       driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
```
